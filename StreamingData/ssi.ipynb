# API


```python
from ssi_fc_data import fc_md_client, model

config = Config()

client = fc_md_client.MarketDataClient(config)
```

```python
def md_get_securities_list():
    req = model.securities('HOSE', 1, 100)
    result = client.securities(config, req)
    return result

def md_get_securities_details():
    req = model.securities_details('HNX', 'FPT', 1, 100)
    print(client.securities_details(config, req))
```

```python
def get_symbol(market):
    req = model.securities(market, 1, 1000)
    result = client.securities(config, req)
    symbol = [item['Symbol'] for item in result['data']]
    return symbol
```

```python
def get_symbol_list():
    hnx = get_symbol('HNX')
    hose = get_symbol('HOSE')
    upcom = get_symbol('UPCOM')
    symbol_list = hnx + hose + upcom
    return symbol_list
```
